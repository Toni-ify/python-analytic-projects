## Exploratory Data Analysis of Supermarket Store Branches

```python
# Import the necessary Modules/Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Dataframe file
df = pd.read_csv('/content/drive/MyDrive/PYTHON PROJECTS/Stores.csv')
df
```

### Data Inspection, Cleaning and Manipulation

```python
# Check Data Info
df.info()
```

```python
# Check for Null/Empty Values.
df.isnull().sum()
```

```python
# Define functions
def convert_to_string(data):
  '''This function converts any value to string dtype.'''
  return f'Store Code {data}'
def convert_to_string1(data):
  '''This function converts any value to string dtype.'''
  return f'Store Area Code {data}'
```

```python
# Create a new column called 'Store_Code'.
df['Store_Code'] = df['Store ID '].apply(convert_to_string)
df['Store_Area_Code'] = df['Store_Area'].apply(convert_to_string1)
df
```

```python
df.info()
```

### Bivariate and Multivariate Analysis

```python
# Check columns.
df.columns
```

```python
# INSIGHT: Top Ten Store area with the most Sales.
store_area_mostSales = df.loc[:, ['Store_Area_Code',  'Store_Sales']].sort_values('Store_Sales', ascending=False).head(10)
store_area_mostSales
```

```python
# Data visualization
plt.figure(figsize=(7, 5))
plt.title(label='Top Ten Store area with the most Sale')
ax = sns.barplot(data=store_area_mostSales,
            y='Store_Area_Code',
            x='Store_Sales',
            orient= 'h',
            # color='grey'
            )
plt.show()
```

#### Observation
Store Area with code 1989 has the most Sales compared to other store areas.

```python
# INSIGHT: Top Ten Store area with the most Sales.
store_area_leastSales = df.loc[:, ['Store_Area_Code',  'Store_Sales']].sort_values('Store_Sales', ascending=True).head(10)
store_area_leastSales
```

```python
# Data visualization
plt.figure(figsize=(7, 5))
plt.title(label='Top Ten Store area with the Least Sale')
ax = sns.barplot(data=store_area_leastSales,
            y='Store_Area_Code',
            x='Store_Sales',
            orient= 'h',
            # color='grey'
            )
plt.show()
```

#### Observation.
The Store Area with code 1250 has the least Sales compared to other Store Areas.

```python
# INSIGHT: Store Area with the Most Daily Customer Count
most_customer_count = df.loc[:, ['Store_Area_Code', 'Daily_Customer_Count']].sort_values('Daily_Customer_Count', ascending=False).head(4)
most_customer_count
```

```python
# Data Visualization.
# Set up the Seaborn style
sns.set(style='whitegrid')

# Create a pie chart using Seaborn
plt.figure(figsize=(6, 4))
plt.pie(most_customer_count['Daily_Customer_Count'],
        labels=most_customer_count['Store_Area_Code'],
        colors=sns.color_palette('pastel'),
        autopct='%1.1f%%',
        shadow=True,
        startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Top four Store Area with the Most Daily Customer Count')
plt.show()
```

#### Observation
Store Area with code 919 has the most Daily Count of Customers.

### Bivariate and Multivariate Analysis

```python
# INSIGHT: States with the most county
# fisrt code:
states_with_most_county = df.groupby(['STNAME'], as_index=False)['CTYNAME'].size().sort_values('size', ascending=False)
states_with_most_county
```

```python
# Continuation of states_with_most_county.
# # Second code:
# top_ten_state = states_with_most_county.groupby(['STNAME'], as_index=False).count().sort_values('count', ascending=False)
# top_ten_state
```

```python
# Data visualization
plt.figure(figsize=(12, 8))
plt.title(label='States with the Most County')
ax = sns.barplot(data=states_with_most_county,
            y='STNAME',
            x='size',
            orient= 'h',
            # color='grey'
            )
plt.show()
```

#### Observations.
Texas State has the highest number of County compared to other States. However, District of Columbia State has the least number of County.

```python
# INSIGHT: Top Three Most Populous States in 2010 Census.
top_populous_state = df.loc[:,['STNAME', 'CTYNAME', 'CENSUS2010POP']].sort_values('CENSUS2010POP', ascending=False).head(3)
top_populous_state
```

```python
# Alternative to above code.
# INSIGHT: Top Three Most Populous States in 2010 Census.
top_populous_state = df.groupby(['STNAME', 'CTYNAME'], as_index=False)['CENSUS2010POP'].max().sort_values('CENSUS2010POP', ascending=False).head(3)
top_populous_state
```

```python
# Data Visualization.
# Set up the Seaborn style
sns.set(style='whitegrid')

# Create a pie chart using Seaborn
plt.figure(figsize=(6, 4))
plt.pie(top_populous_state['CENSUS2010POP'],
        labels=top_populous_state['STNAME'],
        colors=sns.color_palette('pastel'),
        autopct='%1.1f%%',
        shadow=True,
        startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Top Three Most Populous States in 2010 Census')
plt.show()
```

#### Observations
California State has the most populous County with a value of 45.6%.

```python
'''
The key for REGION is as follows:
1 = Northeast
2 = Midwest
3 = South
4 = West
'''
# Define a function
def replace_num(num):
  '''This function takes an int and then returns a string'''
  if num == 1:
    return 'Northeast'
  elif num == 2:
    return 'Mideast'
  elif num == 3:
    return 'South'
  else:
    return 'West'
```

```python
# Create new column/fields called New_region
df['New_region'] = df['REGION'].apply(replace_num)
df
```

```python
# INSIGHT: Region with the most division.
region_divvision = df.groupby('New_region', as_index=False)['DIVISION'].size().sort_values('size', ascending=False)
region_divvision
```

```python
# Alternative to code above.
# INSIGHT: Region with the most division.
region_divvision = df.groupby('New_region', as_index=False)['DIVISION'].count().sort_values('DIVISION', ascending=False)
region_divvision
```

```python
# Data visualization
plt.figure(figsize=(6, 4))
plt.title(label='Region with the Most Division')
ax = sns.barplot(data=region_divvision,
            y='New_region',
            x='DIVISION',
            orient= 'h',
            # color='grey'
            )
plt.show()
```

#### Observation
South Region has the highest most Divisions, compared to Northeast Region with the least Divisions.
